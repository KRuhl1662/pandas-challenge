```python
# Import dependencies

import pandas as pd
```

```python
# Load in csv and store in Pandas dataframe

file_to_load = "Resources/purchase_data.csv"

store_data_df = pd.read_csv(file_to_load)
store_data_df.head()
```

```python
# Check for missing data

store_data_df.count()
```

```python
# Calculate the total number of players
Total_Players = len(store_data_df["SN"].unique())

#Create new dataframe of Total Players
Total_Players_df = pd.DataFrame({"Total Players":[Total_Players]})

Total_Players_df
```

```python
# Basic Calculations (Total Unique Items, Average Price of Item, Total Sales)

# Get total number of unique items using len and .unique
Total_Unique_Items = len(store_data_df["Item ID"].unique())

# Use .mean to find the average price
Average_Price = store_data_df["Price"].mean()


# Use len to find how many items were purchased
Total_Purchased_Items = len(store_data_df["Purchase ID"])

#Use .sum to find total amount of sales
Tot_Revenue = store_data_df["Price"].sum()

#Create a new dataframe of the calculated values to display a summary 
Summary_df = pd.DataFrame({"Number of Unique Items":[Total_Unique_Items],"Average Price":Average_Price,"Number of Purchases":Total_Purchased_Items,"Total Revenue":Tot_Revenue})

# Format
Summary_df["Average Price"] = Summary_df["Average Price"].astype(float).map("${:,.2f}".format)
Summary_df["Total Revenue"] = Summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)    

Summary_df
```

```python
# Age Demographics

#Create bins
bins = [0,9,14,19,24,29,34,39,100]
#Create bin labels
age_groups = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
# Cut store data into bins and then create dataframe (originally did this on just Players but need it for more than one analysis)

store_data_df["Age_Demo"] = pd.cut(store_data_df["Age"], bins, labels = age_groups)
```

```python
# Top Spenders - Find most purchases, the average purchase price, total value

# Group by SN 
Individual_Purchasers = store_data_df.groupby(["SN"])

# Calculate total purchases by individuals
Individ_Purch = Individual_Purchasers["Purchase ID"].count()

#Get Sum of money spent by each individual
Individ_Total_Purch = Individual_Purchasers["Price"].sum()

#Get Average Purchase Price by each individual
Individ_Avg_Purch_Price = Individ_Total_Purch / Individ_Purch

#Create new dataframe to display data
Individ_Analysis = pd.DataFrame({"Purchase Count":Individ_Purch,"Average Purchase Price": Individ_Avg_Purch_Price,"Total Purchase Value":Individ_Total_Purch})

#Format
Individ_Analysis["Average Purchase Price"] = Individ_Analysis["Average Purchase Price"].astype(float).map("${:,.2f}".format)
Individ_Analysis["Total Purchase Value"] = Individ_Analysis["Total Purchase Value"].astype(float).map("${:,.2f}".format)

#Sort
Sort_Individ_Analysis = Individ_Analysis.sort_values(by="Purchase Count",ascending=False)

Sort_Individ_Analysis.head()
```

```python
# Most Popular Items

Pop_Items = store_data_df.loc[:,["Item ID","Item Name","Price"]]
Pop_Items.head()
```

```python
# Group items by Item ID and Item Name
Item_Group = Pop_Items.groupby(["Item ID","Item Name"])

# Calculate the total amount of same item ID/Name sold
Items_Sold = Item_Group["Price"].count()

# Calculate the total amount of money made from sales of same item ID/Name
Tot_Item_Price = Item_Group["Price"].sum()

# Calculate the price per item of same item ID/Name
Item_Price = Tot_Item_Price / Items_Sold
```

```python
# Create new dataframe to display data
Item_Summary = pd.DataFrame({"Purchase Count":Items_Sold,"Item Price":Item_Price, "Total Purchase Value":Tot_Item_Price})

#Sort
By_Purchase = Item_Summary.sort_values(by = "Purchase Count", ascending = False)

# Format
By_Purchase["Item Price"] = By_Purchase["Item Price"].astype(float).map("${:,.2f}".format)
By_Purchase["Total Purchase Value"] = By_Purchase["Total Purchase Value"].astype(float).map("${:,.2f}".format)

By_Purchase.head()
```
